```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
loan_dataset = pd.read_csv('/content/train_u6lujuX_CVtuZ9i (1).csv')
```

```python
loan_dataset.head(5)
```

```python
loan_dataset.isnull().sum()
```

```python
loan_dataset = loan_dataset.dropna()
loan_dataset.isnull().sum()
```

```python
loan_dataset.replace({'Loan_Status':{'N':0,'Y':1}},inplace=True)
```

```python
loan_dataset.head(3
                  )
```

```python
loan_dataset['Dependents'].value_counts()
```

```python
loan_dataset = loan_dataset.replace(to_replace='3+', value = 4 )
```

```python
loan_dataset['Dependents'].value_counts()
```

```python
import plotly.express as px

# Assuming 'loan_dataset' is a Pandas DataFrame
fig = px.histogram(loan_dataset, x='Education', color='Loan_Status')  # Use 'color' instead of 'hue'
fig.show()
```

```python
sns.countplot(loan_dataset, x='Education', hue='Loan_Status')  # Use 'color' instead of 'hue'
```

```python
# convert categorical columns to numerical values
loan_dataset.replace({'Married':{'No':0,'Yes':1},'Gender':{'Male':1,'Female':0},'Self_Employed':{'No':0,'Yes':1},
                      'Property_Area':{'Rural':0,'Semiurban':1,'Urban':2},'Education':{'Graduate':1,'Not Graduate':0}},inplace=True)
```

```python
loan_dataset.head()
```

```python
X = loan_dataset.drop(columns=['Loan_ID','Loan_Status'],axis = 1)
Y = loan_dataset['Loan_Status']
```

```python
print(X)
print(Y)
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y, test_size = 0.1, stratify = Y, random_state = 2)
```

```python
model = svm.SVC(kernel='linear')
```

```python
model.fit(X_train,Y_train)
```

```python
X_train_predict = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_predict,Y_train)

Y_test_predict = model.predict(X_test)
test_data_accuracy = accuracy_score(Y_test_predict,Y_test)
print('Accuracy score of training data : ', training_data_accuracy)
print('Accuracy score of test data : ', test_data_accuracy)
```
